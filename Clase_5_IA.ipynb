**VP, VN, FP y FN**.

Verdaderos positivos (VP) son los casos que sí eran positivos y el modelo los marcó correctamente como positivos. Verdaderos negativos (VN) son los que sí eran negativos y el modelo los reconoció como negativos. Falsos positivos (FP) son negativos que el modelo confundió y etiquetó como positivos, es decir, “falsas alarmas”. Falsos negativos (FN) son positivos que el modelo dejó pasar como si fueran negativos, es decir, “se le escaparon”. Con estos cuatro conteos se construyen prácticamente todas las métricas.

**Recall o sensibilidad.**

Mide qué proporción de los positivos reales logra capturar el modelo. Es especialmente importante cuando perder un positivo es costoso: por ejemplo, no detectar un fraude, una falla crítica o una enfermedad. Un recall alto significa que casi ningún positivo se te está yendo como FN; un recall bajo indica que el modelo está “dejando pasar” demasiados casos que sí eran positivos.

**Precisión (precision).**

Mide qué tan “limpias” son tus predicciones positivas, es decir, de todo lo que marcaste como positivo, qué porcentaje realmente lo era. Es clave cuando una falsa alarma sale cara: por ejemplo, escalar un incidente que no lo es, llamar a un cliente por error o activar un protocolo innecesario. Una precisión alta indica pocos FP; una precisión baja sugiere que estás marcando demasiados negativos como si fueran positivos.

image.png

**Exactitud (accuracy).**

Indica el porcentaje total de aciertos del modelo sumando VP y VN sobre todos los casos. Es intuitiva y útil cuando las clases están balanceadas y el costo de equivocarse es parecido en ambos lados. Sin embargo, puede ser engañosa en escenarios desbalanceados: un modelo puede tener una accuracy alta simplemente porque acierta casi todos los negativos, mientras ignora la clase minoritaria.

image.png

**Especificidad (specificity).**

Evalúa qué proporción de los negativos reales reconoce bien el modelo como negativos. Complementa al recall porque mira el “otro lado”: qué tan bien evitas las falsas alarmas. Una especificidad alta significa pocos FP; una baja indica que estás etiquetando demasiados negativos como positivos.

image.png

**F1 score.**

Combina precisión y recall en un solo número equilibrado. Solo sube cuando ambas están razonablemente altas; si una de las dos cae, el F1 también baja. Es útil para comparar modelos cuando quieres un balance entre “no perder positivos” y “no disparar falsas alarmas”, especialmente en datos desbalanceados.

image.png

**F2 score.**

Es una variante del F1 que otorga más peso al recall que a la precisión. Se usa cuando es mucho peor dejar pasar un positivo (FN) que dar una falsa alarma (FP). En términos prácticos, favorece modelos que recuperan casi todos los positivos aunque eso implique aceptar más falsas alarmas.

image.png

**Fβ (F-beta).**

Generaliza la idea de F1 y F2 permitiéndote ajustar cuánta importancia das al recall frente a la precisión según el problema. Con β mayor que 1 priorizas el recall; con β menor que 1 priorizas la precisión. Esto permite alinear la métrica con el costo real de los errores en tu caso de uso.
